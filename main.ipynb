Импортирование нужых модулей

```python
# Вариант 3 Темник

import openpyxl #openpyxl
import matplotlib.pyplot as plt #matplotlib
from openpyxl import load_workbook
%matplotlib inline
```

<br>Открытие таблицы students.xlsx в папке data в переменную <code>wb</code> <br>
Используем переменную <code>ws</code> для работы с таблицей

```python
wb = load_workbook('data/students.xlsx')
ws = wb.active
```

<br>Подсчет количества оценок выше пяти по математике путем перебора столбца с данными об предмете и запись их числа в переменную <code>math_count</code>

```python
#Переменные, отвечающие за целевой столбец и строку для получения данных из нужной ячейки
column = 'B' 
row = 1

math_count = 0  #Количество оценок выше пяти по математике

print('Оценки выше пяти по математике:')
while (row < 108):  #Перебор столбца 
    row += 1
    if ((ws[column + str(row)].value) > 5):
        math_count += 1
        print((ws[column + str(row)].value), end=" ")
```

<br>Подсчет количества оценок выше пяти по физике путем перебора столбца с данными об предмете и запись их числа в переменную <code>phys_count</code>

```python
#Переменные, отвечающие за целевой столбец и строку для получения данных из нужной ячейки
column = 'C' 
row = 1

phys_count = 0  #Количество оценок выше пяти по физике

print('Оценки выше пяти по физике:')
while (row <= 107):
    row += 1
    if ((ws[column + str(row)].value) > 5):
        phys_count += 1
        print((ws[column + str(row)].value), end=" ")
```

<br>Подсчет количества оценок выше пяти по информатике путем перебора столбца с данными об предмете и запись их числа в переменную <code>inform_count</code>

```python
#Переменные, отвечающие за целевой столбец и строку для получения данных из нужной ячейки
column = 'D' 
row = 1

inform_count = 0  #Количество оценок выше пяти по информатике

print('Оценки выше пяти по информатике:')
while (row <= 107):
    row += 1
    if ((ws[column + str(row)].value) > 5):
        inform_count += 1
        print((ws[column + str(row)].value), end=" ")
```

<br>Используя данные, полученные выше, строим столбчатую диаграмму, в которой отображается количество оценок выше пяти по трем предметам <br>
Она состоит по оси X из трех предметов, а по оси Y из оценок выше пяти по ним.

```python
subjects = ["Математика", "Физика", "Информатика"]  #Предметы
notes = [math_count, phys_count, inform_count]  #Оценки
plt.bar(subjects, notes)    #Построение диаграммы
plt.title("Кол-во учащихся с оценкой выше 5 по предметам")  #Название
plt.xlabel("Предмет")   #Ось X
plt.ylabel("Количество")    #Ось Y
```
